```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../results_all.csv')
```

```python
small = data[data['Text Size'].str.contains('1.6kb')]
```

```python
small[small['Pattern Size'] == 8].plot(kind = 'bar', y = 'Time', x = 'Algorithm',\
                                               rot = 45, figsize = (10, 8), color = 'r', \
                                               title = "DNA - Small Needle & Small Haystack")
plt.show()
```

```python
small[small['Pattern Size'] == 1024].plot(kind = 'bar', y = 'Time', x = 'Algorithm',\
                                               rot = 45, figsize = (10, 8), color = 'g', \
                                               title = "DNA - Big Needle & Small Haystack")
plt.show()
```

```python
big = data[data['Text Size'].str.contains('4.6Mb')]
```

```python
big[big['Pattern Size'] == 8].plot(kind = 'bar', y = 'Time', x = 'Algorithm',\
                                               rot = 45, figsize = (10, 8), color = 'b', \
                                               title = "DNA - Small Needle & Big Haystack")
plt.show()
```

```python
big[big['Pattern Size'] == 1024].plot(kind = 'bar', y = 'Time', x = 'Algorithm',\
                                               rot = 45, figsize = (10, 8), color = 'y', \
                                               title = "DNA - Big Needle & Big Haystack")
plt.show()
```
