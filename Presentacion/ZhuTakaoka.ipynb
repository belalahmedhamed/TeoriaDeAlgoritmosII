# ALGORITMO DE ZHU-TAKAOKA

Este algoritmo sirve para resolver el problema de String Matching.

Es una variaci√≥n del algoritmo de Boyer-Moore.

### Como funciona?

Primero tenemos que alinear el texto y el patron a izquierda.

Luego los recorremos de derecha a izquierda, y en caso de matchear o encontar un miss, se usan los siguientes criterios para shiftear el patron:

"The Good Suffix Rule" (El mismo que en Boyer-Moore)

"2-Substring Rule" (Reemplazando "The Bad Character Rule")

## 2-substring rule

Consiste en encontrar las ocurrencias de Tk y Tk+1 en el patron, y alinearlas el texto.

## Como hacemos para saber si 2-Substrings se encuentran o no en el patron?

Zhu-Takaoka crearon una matriz que guarda las ocurrencias de los substrings.

Consiste en computar para cada par de caracteres (a,b) pertenecientes al alfabeto, la aparicion mas a la derecha en el patron entre las posiciones [0, ... , m-2]

ztBc[a,b] = k con k <= m-2, si x[m-k, m-k+1] es la aparicion mas a la derecha.

```python
Image("img/case1.png")
```

ztBc[a,b] = k con k = m-1, si x[0]=b y (a,b) no aparecen en x[0,...,m-2]

```python
Image("img/case2.png")
```

ztBc[a,b] = k con k <= m-2, si x[0]!=b y (a,b) no aparecen en x[0,...,m-2]

## Good Suffix Rule

Supongamos que ocurre un miss entre el caracter x[i]=a del patron y el caracter y[i+j]=b del texto.

Entonces definimos, x[i+1 .. m-1] = y[i+j+1 .. j+m-1] = u. 

La regla consiste en alinear el segmento u con su proxima ocurrencia mas a la derecha, que no este seguida por a (si es que existiese).

```python
Image("img/example2.png")
```

En caso de no existir, consiste en alinear el mayor v entre y[i+j+1 ... j+m-1] con un match en x.
